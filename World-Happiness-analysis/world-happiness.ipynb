```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#aranging the environment

sns.set_style('darkgrid') #setting the background style
plt.rcParams['figure.figsize']=(15,5) #setting the figure size
plt.rcParams['font.size']=15 #setting the font size
plt.rcParams['font.family']='serif' # setting the font family
plt.rcParams['figure.facecolor']='#40E0D0' #setting the background color
```

```python
#loading the data

dt=pd.read_csv('world-happiness-report-2021.csv')
df=dt.copy()
df.head()
```

```python
#rename the columns
df.rename({"Country name":"Country_Name","Regional indicator":"Regional_Indicator","Ladder score":"Happiness score"},inplace=True,axis=1)
df.head()
```

```python
#exploring the data

df.info()
```

```python
df.describe()
```

```python
#ignoring the columns that are not required
df.drop(['Standard error of ladder score','upperwhisker','lowerwhisker','Ladder score in Dystopia','Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption','Dystopia + residual'],axis=1,inplace=True)
df.head()
```

```python
# understanding correlation between the variables
df.corr()
```

```python
#plotting 10 happiest countries
plt.figure(figsize=(15,5))

a=df.sort_values(by="Happiness score",ascending=False).head(10)
px.line(a,x='Country_Name',y='Happiness score',color='Regional_Indicator',title='10 Happiest countries')
```

```python
#Corrupion in regions 
px.scatter_geo(df,locations='Country_Name',locationmode='country names',color='Regional_Indicator',hover_name='Regional_Indicator',size='Perceptions of corruption',projection='natural earth',title='Corruption in regions')
```

```python

#10 most happiest countries and 10 most unhappiest countries in ters of life expectancy

hppy=df.sort_values(by="Healthy life expectancy",ascending=False).head(10)
uhppy=df.sort_values(by="Healthy life expectancy",ascending=False).tail(10) 


fig,axes=plt.subplots(1,2,figsize=(19,6)) #creating subplots

plt.tight_layout(pad=2) #adjusting the layout

xlabels=hppy["Country_Name"]
xlabels2=uhppy["Country_Name"]

axes[0].set_title("10 happiest countries according to life expenctancy",fontsize=20)
sns.lineplot(y=hppy["Healthy life expectancy"],x=hppy["Country_Name"],ax=axes[0],palette="mako")
axes[0].set_xlabel("Happiness score",fontsize=15)
axes[0].set_ylabel("Country name",fontsize=15)
axes[0].set_xticklabels(xlabels,rotation=30,ha="right",fontsize=15)

axes[1].set_title("10 unhappiest countries according to life expenctancy",fontsize=20)
sns.lineplot(y=uhppy["Healthy life expectancy"],x=uhppy["Country_Name"],ax=axes[1],palette="mako")
axes[1].set_xlabel("Happiness score",fontsize=15)
axes[1].set_ylabel("Country name",fontsize=15)
axes[1].set_xticklabels(xlabels2,rotation=30,ha="right",fontsize=15)

```

```python
#Filling the outliers with meean in the happiness score column:
df_happy=df["Happiness score"]

Q1=df_happy.quantile(0.25)
Q3=df_happy.quantile(0.75)
IQR=Q3-Q1 #interquartile range

df_happy.head()

```

```python
lower_limit=Q1-1.5*IQR #lower limit
upper_limit=Q3+1.5*IQR #upper limit

lower_limit,upper_limit
```

```python
((df_happy>upper_limit) | (df_happy<lower_limit)).sum() #finding the number of outliers
```

```python
#filling the outliers with mean

outlier_lw=(df_happy<lower_limit)
outlier_upp=(df_happy>upper_limit)


df_happy[outlier_lw]=df_happy.mean()  
df_happy[outlier_upp]=df_happy.mean() 

df["Happiness score"]=df_happy  # replacing the happiness score column with the new values

px.box(df,y="Happiness score",color="Regional_Indicator",title="Happiness score of each region") #plotting the boxplot again to check if the outliers are removed
```
